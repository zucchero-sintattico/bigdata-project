```scala
// Print all file names in the directory

def listFilesInDirectory(path: String): Unit = {
  val directory = new File(path)
  if (directory.exists && directory.isDirectory) {
    val files = directory.listFiles.filter(_.isFile)
    files.foreach(file => println(file.getName))
  } else {
    println(s"Il percorso $path non è valido o non è una directory.")
  }
}

println(path_to_datasets)
```
